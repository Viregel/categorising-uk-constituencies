```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
uk_2019 = pd.read_csv("UK-2019.csv")
#names = cols

uk_2019.head()
```

```python
#Remove redundant columns

cols = ["ons_id", "constituency_name","region_name","country_name","electorate","con","lab","ld","brexit","green","snp","pc","dup","sf","sdlp","uup","alliance","all_other_candidates","of_which_other_winner"]

uk_2019 = uk_2019[cols]
uk_party_names = list(uk_2019.iloc[:,5:].columns)
```

```python
#Seperate by country

scotland_2019 = uk_2019[uk_2019["country_name"]=="Scotland"]
england_2019 = uk_2019[uk_2019["country_name"]=="England"]
wales_2019 = uk_2019[uk_2019["country_name"]=="Wales"]
ni_2019 = uk_2019[uk_2019["country_name"]=="Northern Ireland"]

#Take out redundant parties
scotland_2019 = scotland_2019.drop(columns=["pc", "dup", "sf", "sdlp", "uup", "alliance"])
england_2019 = england_2019.drop(columns=["snp", "pc", "dup", "sf", "sdlp", "uup", "alliance"])
wales_2019 = wales_2019.drop(columns=["snp", "dup", "sf", "sdlp", "uup", "alliance"])

new = ni_2019["all_other_candidates"] + ni_2019["con"]
ni_2019 = ni_2019.drop(columns=["all_other_candidates"])
ni_2019["all_other_candidates"] = new

ni_2019 = ni_2019.drop(columns=["con", "lab", "ld", "brexit", "snp", "pc"])
ni_2019
```

```python
def get_non_voters(df):
    new_df = df.copy()
    index = df.columns.get_loc('electorate') + 1
    new_df["non_voting"] = new_df["electorate"] - df.iloc[:,index:].sum(axis='columns')
    return new_df
```

```python
def get_percent_vote(df):
    new_df = df.copy()
    index = df.columns.get_loc('electorate') + 1

    for col in range(len(df.iloc[:,index:].columns)):
        new_df.iloc[:,col+index] = list(round((df.iloc[:,col+index]/df["electorate"])*100, 2))

    return new_df
```

```python
wales_2019 = get_non_voters(wales_2019)
```

```python
get_percent_vote(wales_2019)
```

```python
#Totals

def get_total(df):
    names = []
    totals = []
    index = df.columns.get_loc('electorate')

    for col in range(len(df.iloc[:,index:].columns)):
        names.append(df.columns[index+col])
        totals.append(df.iloc[:,index+col].sum())

    new_df = pd.DataFrame([totals], columns=names)
    return new_df
```

```python
get_total(ni_2019)
```
